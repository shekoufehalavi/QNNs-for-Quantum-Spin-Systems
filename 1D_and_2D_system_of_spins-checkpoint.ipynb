

چیزی که مقاله ی اصلی در نظر نگرفته است، این است که در ضریب اندرکنش بین دو اسپین
در سیستم من یعنی ضرایب هامیلتونی، فاصله باید لحاظ شود. یعنی تاثیر هر اسپینی روی اسپین همسایه اش قوی تر از اسپینهای دیگر باشد، این اندرکنش ها را واقعی تر می کند. من دو حالت یک بعدی(زنجیره ای از اسپینها) و دو بعدی(شبکه ای از اسپینها) را در نظر گرفته ام. برای مورد اول، یعنی چینش یک بعدی از اسپینها باید فرمولی برای اندرکنش بین اسپینها در جهت ایکس و زد در نظر گرفت که وابستگی به فاصلخ در ان وجود داشته باشد. این وابستگی را میتوان خطی در نظر گرفت یا توانی. یعنی اندرکنش بین دواسپین با فاصله بطور نمایی افت کند یا خطی

```python
# Consider exponencial decay

import numpy as np
num_qubits = 6
matrix_size = 2 ** num_qubits
num_repetitions = 3
Mz_values = np.zeros((num_repetitions, matrix_size))

eigenvalues = []
eigenvectors = []

# Parameters for distance-dependent interactions
J0 = 1.0
K0 = 1.0
lambda_J = 1.0  # Decay length for J
lambda_K = 1.0  # Decay length for K

# Function to calculate distance-dependent interactions
def distance_dependent_interaction(d, J0, lambda_):
    return J0 * np.exp(-d / lambda_)

# 1D case: distance is |i - j|
def get_distance_1d(i, j):
    return abs(i - j)

for repetition in range(num_repetitions):
    # Create random numbers for external fields
    np.random.seed(repetition + 20)
    h = np.random.uniform(low=-0.04, high=0.04)

    np.random.seed(repetition + 30)
    g = np.random.uniform(low=-6, high=6, size=num_qubits)

    # Create distance-dependent interaction matrices J and K
    J = np.zeros((num_qubits, num_qubits))
    K = np.zeros((num_qubits, num_qubits))

    for i in range(num_qubits):
        for j in range(num_qubits):
            if i < j:
                distance = get_distance_1d(i, j)
                J[i, j] = distance_dependent_interaction(distance, J0, lambda_J)
                K[i, j] = distance_dependent_interaction(distance, K0, lambda_K)

    # Symmetrize the matrices
    J = J + J.T
    K = K + K.T

    # Create the Hamiltonian matrix H
    matrix = np.zeros((matrix_size, matrix_size))
    for i in range(matrix_size):
        for j in range(matrix_size):
            H_value = 0
            for i1 in range(num_qubits):
                for j1 in range(num_qubits):
                    if j == i ^ (2 ** i1 + 2 ** j1):
                        H_value += K[i1, j1]
                    if j == i:
                        sign = 1
                        if (i & 2 ** i1) != 0:
                            sign = -sign
                        if (i & 2 ** j1) != 0:
                            sign = -sign
                        H_value += sign * J[i1, j1]
            for i1 in range(num_qubits):
                if j == i ^ (2 ** i1):
                    H_value += g[i1]
            if j == i:
                H_value += h
            matrix[i, j] = H_value

    # Calculate eigenvalues and eigenvectors
    eigenval, eigenvect = np.linalg.eigh(matrix)
    min_eigenval = np.min(eigenval)
    min_eigenvec = eigenvect[:, np.argmin(eigenval)]

    eigenvalues.append(min_eigenval)
    eigenvectors.append(min_eigenvec)

# Save the eigenvectors as a CSV file
eigenvectors_csv = np.array(eigenvectors)
np.savetxt('eigenvectors.csv', eigenvectors_csv, delimiter=',')

# Print eigenvalues and eigenvectors for each repetition
for repetition in range(num_repetitions):
    print("Repetition", repetition + 1)
    print("Eigenvalue:", eigenvalues[repetition])
    print("Eigenvector:", eigenvectors[repetition])
    print()

```

```python
#creating the matrix of data, eigenvectors. each column is eigenvector of a certain H.
eigenvectors_matrix = np.column_stack(eigenvectors)

print("Eigenvectors Matrix:")
print(eigenvectors_matrix)
```

```python

Mzt = []
for column in eigenvectors_matrix.T:
    i = 0
    Mz = 0
    for component in column:
        for n_prime in range(num_qubits):
            if (2 ** n_prime) & i != 0 :
                Mz += abs(component) ** 2  / num_qubits
          #Mz += abs(component) ** 2 * (2 * ((2 ** n_prime) & i) - 1) / num_qubits
        i += 1
    Mz=2*Mz-1
    Mzt.append(Mz)

Mzt_row = np.array(Mzt)
print("Mzt:", Mzt,"\n")
print(len(Mzt),"\n")
print(Mz)
```

```python
import numpy as np

num_qubits_per_side = 3  # For a 3x3 grid
num_qubits = num_qubits_per_side ** 2
matrix_size = 2 ** num_qubits
num_repetitions = 3
Mz_values = np.zeros((num_repetitions, matrix_size))

eigenvalues = []
eigenvectors = []

# Parameters for distance-dependent interactions
J0 = 1.0
K0 = 1.0
lambda_J = 1.0  # Decay length for J
lambda_K = 1.0  # Decay length for K

# Function to calculate distance-dependent interactions
def distance_dependent_interaction(d, J0, lambda_):
    return J0 * np.exp(-d / lambda_)

# 2D case: distance is sqrt((i1 - j1)^2 + (i2 - j2)^2)
def get_distance_2d(pos1, pos2):
    return np.sqrt((pos1[0] - pos2[0]) ** 2 + (pos1[1] - pos2[1]) ** 2)

# Map 1D index to 2D position
def index_to_2d(index, side_length):
    return (index // side_length, index % side_length)

for repetition in range(num_repetitions):
    # Create random numbers for external fields
    np.random.seed(repetition + 20)
    h = np.random.uniform(low=-0.04, high=0.04)

    np.random.seed(repetition + 30)
    g = np.random.uniform(low=-6, high=6, size=num_qubits)

    # Create distance-dependent interaction matrices J and K
    J = np.zeros((num_qubits, num_qubits))
    K = np.zeros((num_qubits, num_qubits))

    for i in range(num_qubits):
        for j in range(num_qubits):
            if i < j:
                pos_i = index_to_2d(i, num_qubits_per_side)
                pos_j = index_to_2d(j, num_qubits_per_side)
                distance = get_distance_2d(pos_i, pos_j)
                J[i, j] = distance_dependent_interaction(distance, J0, lambda_J)
                K[i, j] = distance_dependent_interaction(distance, K0, lambda_K)

    # Symmetrize the matrices
    J = J + J.T
    K = K + K.T

    # Create the Hamiltonian matrix H
    matrix = np.zeros((matrix_size, matrix_size))
    for i in range(matrix_size):
        for j in range(matrix_size):
            H_value = 0
            for i1 in range(num_qubits):
                for j1 in range(num_qubits):
                    if j == i ^ (2 ** i1 + 2 ** j1):
                        H_value += K[i1, j1]
                    if j == i:
                        sign = 1
                        if (i & 2 ** i1) != 0:
                            sign = -sign
                        if (i & 2 ** j1) != 0:
                            sign = -sign
                        H_value += sign * J[i1, j1]
            for i1 in range(num_qubits):
                if j == i ^ (2 ** i1):
                    H_value += g[i1]
            if j == i:
                H_value += h
            matrix[i, j] = H_value

    # Calculate eigenvalues and eigenvectors
    eigenval, eigenvect = np.linalg.eigh(matrix)
    min_eigenval = np.min(eigenval)
    min_eigenvec = eigenvect[:, np.argmin(eigenval)]

    eigenvalues.append(min_eigenval)
    eigenvectors.append(min_eigenvec)

# Save the eigenvectors as a CSV file
eigenvectors_csv = np.array(eigenvectors)
np.savetxt('eigenvectors.csv', eigenvectors_csv, delimiter=',')

# Print eigenvalues and eigenvectors for each repetition
for repetition in range(num_repetitions):
    print("Repetition", repetition + 1)
    print("Eigenvalue:", eigenvalues[repetition])
    print("Eigenvector:", eigenvectors[repetition])
    print()
```
