```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split


num_qubits = 6
matrix_size = 2**num_qubits
num_repetitions = 1000
Mz_values = np.zeros((num_repetitions, matrix_size))


eigenvalues = []
eigenvectors = []

for repetition in range(num_repetitions):


    #create random numbers for H.
    np.random.seed(repetition)
    J = np.random.uniform(low=-1, high=0, size=(num_qubits, num_qubits))
    for i in range(num_qubits):
        for j in range(num_qubits):
            if i >= j:
                J[i, j] = 0
    J = (J + J.T)

    np.random.seed(repetition +10)
    K = np.random.uniform(low=-1, high=1, size=(num_qubits, num_qubits))
    for i in range(num_qubits):
        for j in range(num_qubits):
            if i >= j:
                K[i, j] = 0
    K = (K + K.T)

    np.random.seed(repetition +20)
    h = np.random.uniform(low=-0.04, high=0.04)

    np.random.seed(repetition +30)
    g = np.random.uniform(low=-6, high=6, size=num_qubits)




    # create the matrix of H.
    matrix = np.zeros((matrix_size, matrix_size))
    for i in range(matrix_size):
        for j in range(matrix_size):
            H_value = 0
            for i1 in range(num_qubits):
                for j1 in range(num_qubits):
                    if j == i ^ (2 ** i1 + 2 ** j1):
#This part adds contributions to H_value based on the coupling constants K and the XOR operation (^).
#It checks if the bit-flipped indices i ^ (2 ** i1 + 2 ** j1) are equal to j and adds the corresponding K valeu.
                        H_value += K[i1, j1]
                    if j == i:
                        sign = 1
                        if (i & 2 ** i1) != 0:# & returns 1 if both the bits are 1, otherwise 0.
                            sign = -sign
                        if (i & 2 ** j1) != 0:
                            sign = -sign

                        H_value += sign * J[i1, j1]

            for i1 in range(num_qubits):
                if j == i ^ (2 ** i1):# external fields g for bit-flipped indices.
                    H_value += g[i1]

            if j == i:
                H_value += h # external field h when j is equal to i.
            matrix[i, j] = H_value




    # calculate eigenvalues and eigenvectors.
    eigenval, eigenvect = np.linalg.eigh(matrix)
   # print(eigenval)
   # print(eigenvect)
    min_eigenval = np.min(eigenval)
    min_eigenvec = eigenvect[np.argmin(eigenval), :]

    eigenvalues.append(min_eigenval)
    eigenvectors.append(min_eigenvec)

#creating the matrix of data, eigenvectors. each column is eigenvector of a certain H.
eigenvectors_matrix = np.column_stack(eigenvectors)

# Calculate Mz values from eigenvectors_matrix
Mzt = []
for column in eigenvectors_matrix.T:
    Mz = 0
    for i, component in enumerate(column):
        for n_prime in range(num_qubits):
            if (2 ** n_prime) & i != 0:
                Mz += abs(component) ** 2 / num_qubits
    Mz = 2 * Mz - 1
    Mzt.append(Mz)

Mzt_row = np.array(Mzt)
features = np.array(eigenvectors)
Mzt= np.array(Mzt_row)
print(Mzt)
print("minimum Mz:", min(Mzt))
print("maximum Mz:", max(Mzt))

#--------------------------------------------------
# Training NN model to predict Mz for new H
#--------------------------------------------------

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(features, Mzt_row, test_size=0.2, random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1)  # 0.25*0.8=0.2

# Define and train the neural network
tf.random.set_seed(42)
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(60, activation='relu', input_shape=(X_train.shape[1],)))
model.add(tf.keras.layers.Dense(32, activation='sigmoid'))
#model.add(tf.keras.layers.Dense(32, activation='softmax'))
model.add(tf.keras.layers.Dense(32, activation='tanh'))
#model.add(tf.keras.layers.Dense(32, activation='mish'))
model.add(tf.keras.layers.Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
history = model.fit(X_train, y_train, epochs=100, batch_size=16, validation_data=(X_val, y_val))

# Plot training history
pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1)  # Set range to [0-1]
plt.show()

# Evaluate the model
test_loss = model.evaluate(X_test, y_test)
print("Test Loss:", test_loss)

# Save and load the trained model
model.save('trained_model.keras')

#----------------------------------------------------------
# Phase diagram drawing for differant H configurations
#----------------------------------------------------------
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

# Load the trained model from Code 1
model = tf.keras.models.load_model('trained_model.keras')

# Constants
num_qubits = 6
matrix_size = 2 ** num_qubits
num_samples = 10  # Number of K/J and g/J values to plot
num_repetitions = 3  # Averaging over H configurations for each point
h_values = [-0.04, 0, 0.04]  # 3 plots for different h values

K_J_strengths = np.linspace(0, 1, num_samples)  # K/J strength varying from 0 to 1
g_J_strengths = np.linspace(0, 6, num_samples)  # g/J strength varying from 0 to 6
J_values = np.random.uniform(-1, 0, num_repetitions)  # J is generated randomly for each repetition

# Function to predict Mz using the neural network
def predict_Mz(min_eigenvector):
    # Prepare the eigenvector for prediction (reshape to match input format of the NN)
    min_eigenvector = np.array(min_eigenvector).reshape(1, -1)
    Mz_pred = model.predict(min_eigenvector)
    return Mz_pred[0][0]

# Function to calculate exact Mz from the eigenvector
def calculate_exact_Mz(min_eigenvector):
    # Calculaing Mz
    Mz_exact = 0
    for i, component in enumerate(min_eigenvec):
        for n_prime in range(num_qubits):
            if (2 ** n_prime) & i != 0:
                Mz_exact += abs(component) ** 2 / num_qubits
    return 2 * Mz_exact - 1  # Normalize magnetization between -1 and 1


# Scatter plot loop for each h value
for h in h_values:  # Loop for different h values
    Mz_avg_pred = np.zeros((num_samples, num_samples))  # Store predicted Mz averages for each (K/J, g/J) pair
    Mz_avg_exact = np.zeros((num_samples, num_samples))  # Store exact Mz averages for each (K/J, g/J) pair

    # Loop over g values (x-axis)
    for i, g_strength in enumerate(g_J_strengths):
        # Loop over K values (y-axis)
        for j, K_strength in enumerate(K_J_strengths):
            Mz_pred_values = []
            Mz_exact_values = []

            # Repeat for each repetition to average Mz for (K/J, g/J) pair
            for repetition in range(num_repetitions):
                # Generate random J, K matrices and g vector
                J = np.random.uniform(-1, 0, (num_qubits, num_qubits))
                K = np.random.uniform(-K_strength, K_strength, (num_qubits, num_qubits))
                g_vector = np.random.uniform(low=-g_strength, high=g_strength, size=num_qubits)

                # Symmetrize J and K matrices (upper triangular)
                for p in range(num_qubits):
                    for q in range(p):
                        J[p, q] = 0
                        K[p, q] = 0
                J = J + J.T
                K = K + K.T

                # Create Hamiltonian matrix H
                matrix = np.zeros((matrix_size, matrix_size))
                for p in range(matrix_size):
                    for q in range(matrix_size):
                        H_value = 0
                        for i1 in range(num_qubits):
                            for j1 in range(num_qubits):
                                if q == p ^ (2 ** i1 + 2 ** j1):
                                    H_value += K[i1, j1]
                                if q == p:
                                    sign = 1
                                    if (p & (2 ** i1)) != 0:
                                        sign = -sign
                                    if (p & (2 ** j1)) != 0:
                                        sign = -sign
                                    H_value += sign * J[i1, j1]

                        for i1 in range(num_qubits):
                            if q == p ^ (2 ** i1):
                                H_value += g_vector[i1]
                        if q == p:
                            H_value += h
                        matrix[p, q] = H_value

                # Calculate the eigenvalues and eigenvectors
                eigenval, eigenvect = np.linalg.eigh(matrix)
                min_eigenvec = eigenvect[np.argmin(eigenval), :]  # Ground state eigenvector

                # Use the neural network to predict Mz for the minimum eigenvector
                Mz_pred = predict_Mz(min_eigenvec)
                Mz_pred_values.append(Mz_pred)

                # Calculate the exact Mz from the eigenvector
                Mz_exact = calculate_exact_Mz(min_eigenvec)
                Mz_exact_values.append(Mz_exact)

            # Average Mz over repetitions and store in the arrays
            Mz_avg_pred[j, i] = np.mean(Mz_pred_values)
            Mz_avg_exact[j, i] = np.mean(Mz_exact_values)


    # Plot the phase diagram for predicted Mz
    plt.figure(figsize=(8, 6))
    sns.heatmap(Mz_avg_pred, cmap='coolwarm', xticklabels=np.round(g_J_strengths, 2), yticklabels=np.round(K_J_strengths, 2))
    plt.title(f"Phase Diagram (Predicted Mz) for h = {h}")
    plt.xlabel("g/J")
    plt.ylabel("K/J")
    plt.show()

    # Plot the phase diagram for exact Mz
    plt.figure(figsize=(8, 6))
    sns.heatmap(Mz_avg_exact, cmap='coolwarm', xticklabels=np.round(g_J_strengths, 2), yticklabels=np.round(K_J_strengths, 2))
    plt.title(f"Phase Diagram (Exact Mz) for h = {h}")
    plt.xlabel("g/J")
    plt.ylabel("K/J")
    plt.show()

    # Plot the phase diagram for predicted Mz
    plt.figure(figsize=(8, 6))
    sns.heatmap(Mz_avg_pred, cmap='coolwarm', annot=True, fmt=".2f",
                xticklabels=np.round(g_J_strengths, 2), yticklabels=np.round(K_J_strengths, 2))
    plt.title(f"Phase Diagram (Predicted Mz) for h = {h}")
    plt.xlabel("g/J")
    plt.ylabel("K/J")
    plt.show()

    # Plot the phase diagram for exact Mz
    plt.figure(figsize=(8, 6))
    sns.heatmap(Mz_avg_exact, cmap='coolwarm', annot=True, fmt=".2f",
                xticklabels=np.round(g_J_strengths, 2), yticklabels=np.round(K_J_strengths, 2))
    plt.title(f"Phase Diagram (Exact Mz) for h = {h}")
    plt.xlabel("g/J")
    plt.ylabel("K/J")
    plt.show()
# Plot Mz distribution for each h separately
for h in h_values:
    plt.figure(figsize=(10, 6))
    plt.hist(Mz_avg_exact.flatten(), bins=30, alpha=0.7, color='blue', edgecolor='black')
    plt.title(f"Distribution of exact Mz for h = {h}")
    plt.xlabel("Mz")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(10, 6))
    plt.hist(Mz_avg_pred.flatten(), bins=30, alpha=0.7, color='green', edgecolor='black')
    plt.title(f"Distribution of predicted Mz for h = {h}")
    plt.xlabel("Predicted Mz")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()
```
